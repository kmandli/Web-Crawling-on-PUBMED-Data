```python
import nltk
import re
from nltk.tokenize import word_tokenize
from math import *
from gensim.models import KeyedVectors

#model = gensim.models.KeyedVectors.load_word2vec_format(os.path.join(os.path.dirname('C:/Users/kavya/Desktop/md#s/glove.6B.50d.txt'), 'C:/Users/kavya/Desktop/mds/glove.6B.50d.txt'), binary=True)


model = KeyedVectors.load_word2vec_format('C:/Users/kavya/Desktop/mds/pubmed/glove.6B.50d (1).txt', binary= False )
side_effects_file = open('C:/Users/kavya/Desktop/mds/pubmed/first_output.xml', 'r',encoding="utf8")
effects_words_output = open('C:/Users/kavya/Desktop/mds/pubmed/Final_list.xml', 'w',encoding="utf8")

hyperlink_patterns = re.compile(r'(\[\[(\w\W*?)+\]\])')

header_text = "The side effects of {0} are:"
current_header = ""
extracted_text = ""
side_effects_list = set()
for line in side_effects_file:
    if "<title>" in line:
        line = line.replace("<title>", "").replace("</title>", "").strip()
        current_header = header_text
        current_header = current_header.format(line) + "\n"
        
        for (match, text) in re.findall(hyperlink_patterns, extracted_text):
            side_effects_list.add(match[2:-2])
        
        pos_tag_list = set()
        
        pos_text = word_tokenize(extracted_text)
        pos__text_tags = nltk.pos_tag(pos_text)
        
        [pos_tag_list.add(s[0]) for s in pos__text_tags if s[1] == 'NN' or s[1] == 'JJ']
        
        for eachVal in pos_tag_list:
            if eachVal in model.vocab:
                result = model.similarity('nausea', eachVal)
                result1 = model.similarity('itching', eachVal)
                result2 = model.similarity('inflammation', eachVal)
                result3 = model.similarity('diarrhea', eachVal)
                result4 = model.similarity('glaucoma', eachVal)
                result5 = model.similarity('fever', eachVal)
                result6 = model.similarity('constipation', eachVal)
                result7 = model.similarity('metastasis', eachVal)
                result8 = model.similarity('euphoria', eachVal)
                result9 = model.similarity('anxiety', eachVal)
                result10 = model.similarity('sociability', eachVal)
                result11 = model.similarity('pain', eachVal)
                #result12 = model.similarity('talkitiveness', eachVal)
                result13 = model.similarity('paralysis', eachVal)
                result14 = model.similarity('impair', eachVal)
                
                
                if result > 0.55 or result1 > 0.55 or result2 > 0.55 or result3 > 0.55 or result4 > 0.55 or result5 > 0.55 or result6 > 0.55 or result7 > 0.55:
                    side_effects_list.add(eachVal)
        
        current_header = current_header + ",".join(side_effects_list)
        current_header = current_header + '\n\n'
        
        if side_effects_list:
            effects_words_output.write(current_header)
        
        print(current_header)
        extracted_text = ""
        side_effects_list = set()
        
    else:
        extracted_text = extracted_text + " " + line
```
